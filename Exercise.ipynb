#### Exercise 3: Caesar Cipher : Encryption & Decryption

```python
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

def caesar(direction,orginal_text,shift_amount):
    if direction == 'encode':
        encrypt_text=""
        for i in orginal_text:
            if i not in alphabet:
                encrypt_text += i
            else:    
                index_of_i = alphabet.index(i)
                #print(index_of_i)
                shifted_index = index_of_i + shift_amount
                
                shifted_index = shifted_index % len(alphabet)

                #print(shifted_index)
                text = alphabet[shifted_index]
                encrypt_text += text
        print(f"Here is the decoded result: {encrypt_text}")
    elif direction == 'decode':
        decrypt_text=""
        for i in orginal_text:
            if i not in alphabet:
                decrypt_text += i
            else:    
                index_of_i = alphabet.index(i)
                #print(index_of_i)
                shifted_index = index_of_i - shift_amount
                
                shifted_index = shifted_index % len(alphabet)

                #print(shifted_index)
                text = alphabet[shifted_index]
                decrypt_text += text
            
        print(f"Here is the decoded result: {decrypt_text}")
    else:
        print("choose only 'encode' or 'decode' with correct spelling")    

    
should_continue = True
while should_continue:
    direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n").lower()
    text = input("Type your message:\n").lower()
    shift=int(input("Type the shift number:\n"))
    caesar(direction,text,shift)
    repeat = input("Type 'yes' if you want to go again. Otherwise type 'no'\n").lower()
    if repeat == 'no':
        should_continue = False
        print("Goodbye")
```

  A normal year has 365 days, leap years have 366, with an extra day in February. The reason why we have leap years is really fascinating, this video does it more justice


  This is how you work out whether if a particular year is a leap year. 

  - on every year that is divisible by 4 with no remainder

  - except every year that is evenly divisible by 100 with no remainder 

  - unless the year is also divisible by 400 with no remainder   


  If English is not your first language, or if the above logic is confusing, try using this flow chart.


  e.g. The year 2000: 

  2000 Ã· 4 = 500 (Leap)  
  2000 Ã· 100 = 20 (Not Leap)  
  2000 Ã· 400 = 5 (Leap!)  
  So the year 2000 is a leap year. 



  But the year 2100 is not a leap year because: 

  2100 Ã· 4 = 525 (Leap)  
  2100 Ã· 100 = 21 (Not Leap)  
  2100 Ã· 400 = 5.25 (Not Leap)

  Example Input 2

```python
## Code: Leap year or not

def leap_year_or_not(num):
    if num % 4 == 0:
        if num % 100 == 0:
            if num % 400 ==0:
                return True
            else:
                return False  
        else:
            return True
    else:
        return False
    
leap_year_or_not(2100)
```

#### Exercise 10: Check If a String is Palindrome

```python
# one way
def is_palindrome(word):
    word = word.lower().replace(" ", "")
    if word[::-1] == word:
        return True
    else:
        return False
is_palindrome("baby")
```

```python
# different way without slicing:

def is_palindrome(word):
    word = word.lower().replace(" ", "")

    i=0
    j=len(word)-1

    while i<j:
        if word[i] == word[j]:
            i += 1
            j -= 1
        else:
            return False
    return True


is_palindrome("aba")
```

#### Exercise 18: Pyramid Pattern

Pyramid Pattern
Problem Description:

You are given an integer n. Your task is to return a pyramid pattern of '*' where each side has n rows, represented as a list of strings. The pyramid is centered, with 1 star in the first row, 3 stars in the second row, and so on, increasing by 2 stars per row until the base row has 2n - 1 stars.



Input:

A single integer n, where 1 <= n <= 100.



Output:

A list of strings where each string contains stars ('*') centered, forming a pyramid shape. Each row has an increasing number of stars, with appropriate spaces for centering.



Example:

    Input: 3
    Output: ['  *  ', ' *** ', '*****']
    
    Input: 5
    Output: ['    *    ', '   ***   ', '  *****  ', ' ******* ', '*********']

```python
def generate_pyramid(n):
    if n<1 or n>100:
        return "'n' should be >=1 or <=100"
    
    lst = []
    times=1
    for i in range(1,n+1):
        lst.append(f"{(n-i)*' '}{'*'*times}{(n-i)*' '}")
        times+=2

    return lst

generate_pyramid(10)
```

#### Exercise 24: Sandglass Pattern

Sandglass Pattern
Problem Description:

You are given an integer n. Your task is to return a sandglass pattern of '*', where the first row contains 2n - 1 stars and each subsequent row decreases the number of stars by 2, until the last row contains a single star. After reaching the smallest width, the pattern then continues with the same number of stars increasing back to 2n - 1. The stars in each row should be centered.



Input:

A single integer n, where 1 <= n <= 100.



Output:

A list of strings where each string represents a row in the sandglass pattern.



Example:

    Input: 3
    Output: ['*****', ' *** ', '  *  ', ' *** ', '*****']
    
    Input: 4
    Output: ['*******', ' ***** ', '  ***  ', '   *   ', '  ***  ', ' ***** ', '*******']

```python
def generate_sandglass(n):
    if n<1 or n>100:
        return "'n' should be >=1 or <=100"
    
    lst=[]
    num=2*n -1
    for i in range(1,n+1):
        operation=(i-1)*' '
        lst.append(f"{operation}{'*'*num}{operation}")
        num-=2
    
    lst2=lst[:-1][::-1]
    lst.extend(lst2)
    return(lst)

generate_sandglass(10)
```

#### Exercise 52: Decimal to Binary

Decimal to Binary
Problem Description:

You are given an integer n. Your task is to return its binary representation as a string. Do not use any built-in functions for conversion.



Input:

A single integer n, where -10^9 <= n <= 10^9.



Output:

A string representing the binary representation of n.



Example:

Input: n = 5
Output: "101"
 
Input: n = -5
Output: "-101"

```python
def int_to_binary(n):
    """
    Function to convert an integer to its binary representation without using built-in functions.
    
    Parameters:
    n (int): The integer to convert.
    
    Returns:
    str: The binary representation of the integer.
    """
    if n == 0:
        return "0"  # Special case for zero
 
    # Store the binary representation
    binary_representation = ""
    
    # Handle negative numbers
    is_negative = n < 0
    if is_negative:
        n = -n  # Work with the absolute value
 
    # Convert to binary
    while n > 0:
        remainder = n % 2
        binary_representation = str(remainder) + binary_representation  # Prepend the remainder
        n = n // 2  # Update n to be n divided by 2 (floor division)
 
    # Add the negative sign for negative numbers
    if is_negative:
        binary_representation = "-" + binary_representation
 
    return binary_representation
```

```python
def binary_to_decimal(binary_str):
    """
    Function to convert a binary string to its decimal integer representation without using built-in functions.
    
    Parameters:
    binary_str (str): The binary string to convert.
    
    Returns:
    int: The decimal representation of the binary string.
    """
    decimal_value = 0
    length = len(binary_str)
    
    # Iterate over each character in the binary string
    for i in range(length):
        # Get the digit (either '0' or '1')
        digit = binary_str[length - 1 - i]  # Start from the end of the string
        
        # If the digit is '1', add the corresponding power of 2 to the decimal value
        if digit == '1':
            decimal_value += 2 ** i  # Add 2 raised to the current power
 
    return decimal_value
```

Return True if t is an anagram of s, and False otherwise.



Example:

```python
def is_anagram(s,t):
    """
    Function to check if t is an anagram of s.
    
    Parameters:
    s (str): The first input string.
    t (str): The second input string.
    
    Returns:
    bool: True if t is an anagram of s, False otherwise.
    """
    if len(s)!=len(t):
        return False
    
    s_frequency={}
    t_frequency={}

    for i in s:
        if i in s_frequency:
            s_frequency[i]+=1
        else:
            s_frequency[i]=1 
    
    for i in t:
        if i in t_frequency:
            t_frequency[i]+=1
        else:
            t_frequency[i]=1 

    for i in s_frequency:
        if i not in t_frequency:
            return False
        elif s_frequency[i] != t_frequency[i]:
            return False
    
    return True
```

##### Variable Length Arguments in Python

    In Python, functions can accept a variable number of arguments using two special symbols:
    *args (Non-keyword variable-length arguments)
    **kwargs (Keyword variable-length arguments)
    These allow us to handle an arbitrary number of arguments in a function call.

##### 1. *args (Non-keyword Arguments)

    *args allows you to pass a variable number of positional arguments to a function.
    The function receives them as a tuple.

    Example of *args
    def add_numbers(*args):
        total = sum(args)
        return total
    
    print(add_numbers(1, 2, 3))        # Output: 6
    print(add_numbers(5, 10, 15, 20))  # Output: 50
    print(add_numbers())               # Output: 0 (no arguments given)

    Explanation
    The *args parameter collects all extra positional arguments into a tuple.
    The function sum(args) adds them up.


##### 2. **kwargs (Keyword Arguments)

    **kwargs allows you to pass a variable number of keyword arguments to a function.
    The function receives them as a dictionary.

    Example of **kwargs
    def print_info(**kwargs):
        for key, value in kwargs.items():
            print(f"{key}: {value}")
    
    print_info(name="Alice", age=25, city="New York")

    Output
    name: Alice
    age: 25
    city: New York

    Explanation
    The **kwargs parameter collects all extra keyword arguments into a dictionary.
    We iterate over the dictionary using .items() to print the keys and values.

##### Combining *args and **kwargs

    You can use both *args and **kwargs in the same function.

    Example
    def mixed_arguments(a, b, *args, **kwargs):
        print(f"Fixed arguments: a={a}, b={b}")
        print(f"Variable positional arguments (args): {args}")
        print(f"Variable keyword arguments (kwargs): {kwargs}")
    
    mixed_arguments(1, 2, 3, 4, 5, name="Alice", age=25)
    Output
    Fixed arguments: a=1, b=2
    Variable positional arguments (args): (3, 4, 5)
    Variable keyword arguments (kwargs): {'name': 'Alice', 'age': 25}

    Explanation
    a and b are mandatory positional arguments.
    *args collects extra positional arguments into a tuple (3, 4, 5).
    **kwargs collects extra keyword arguments into a dictionary {'name': 'Alice', 'age': 25}.


##### Positional and Keyword Arguments in Python

    When defining a function, parameters can be passed in two ways:
    Positional Arguments
    Keyword Arguments

##### 1. Positional Arguments
    These are arguments that must be passed in the correct order.
    The number of arguments must match the number of parameters in the function definition.

    Example
    def greet(name, age):
        print(f"Hello, my name is {name} and I am {age} years old.")
    
    greet("Alice", 25)  # Correct
    greet(25, "Alice")  # Incorrect order

    Output
    Hello, my name is Alice and I am 25 years old.
    Hello, my name is 25 and I am Alice years old.  # Incorrect

    Explanation
    greet("Alice", 25) works correctly because the order of arguments matches the function parameters.
    greet(25, "Alice") swaps name and age, leading to incorrect output.

##### 2. Keyword Arguments
    These are arguments that are passed with explicit names.
    The order does not matter.
    This improves readability.

    Example
    def greet(name, age):
        print(f"Hello, my name is {name} and I am {age} years old.")
    
    greet(name="Alice", age=25)     # Correct
    greet(age=30, name="Bob")       # Correct (order doesn't matter)

    Output
    Hello, my name is Alice and I am 25 years old.
    Hello, my name is Bob and I am 30 years old.

    Explanation
    The function receives arguments using named parameters, so order does not matter.
    This makes the function more readable.

##### Mixing Positional and Keyword Arguments
    Positional arguments must come before keyword arguments.
    Otherwise, Python will raise an error.

    Example (Correct)
    def describe_pet(animal, name):
        print(f"I have a {animal} named {name}.")
    
    describe_pet("dog", name="Buddy")

    Example (Incorrect)
    describe_pet(animal="dog", "Buddy")  # âŒ SyntaxError
    Error Message
    SyntaxError: positional argument follows keyword argument

##### Default Values in Functions
    You can assign default values to function parameters.
    If an argument is not provided, the default value is used.

    Example
    def greet(name="Guest"):
        print(f"Hello, {name}!")
    
    greet("Alice")  # Output: Hello, Alice!
    greet()         # Output: Hello, Guest!
    Explanation
    If name is not provided, "Guest" is used as the default.

### Linear Search:

The Linear Search algorithm searches through an array and returns the index of the value it searches for, if the value is not present it will return -1.

In simple:

        return the target index if present
        else return -1

#### Linear Search Implementation

To implement the Linear Search algorithm we need:

1. An array with values to search through.
2. A target value to search for.
3. A loop that goes through the array from start to end.
4. An if-statement that compares the current value with the target value, and returns the current index if the target value is found.
5. After the loop, return -1, because at this point we know the target value has not been found.



```python
def linear_search(lst,target):

    for index, val in enumerate(lst):
         if val == target:
              return index
         
    return -1 # if no match for target return -1
```

    try:

    except:
      print("Something else went wrong")

You can use the else keyword to define a block of code to be executed if no errors were raised:

##### Example

    try:
      print("Hello")
    except:
      print("Something went wrong")
    else:
      print("Nothing went wrong")

##### What is an Abstract Base Class?
An Abstract Base Class is a class that cannot be instantiated directly. Instead, it serves as a blueprint for other classes. It defines methods that must be created in any subclass that inherits from it. Think of it as a template for other classes.

##### Why Use Abstract Base Classes?
- Enforce Structure: They ensure that certain methods are implemented in any subclass.
- Provide a Common Interface: They allow you to define a common interface for a group of related classes.

##### Example to Understand Abstract Base Classes
Imagine you have different types of shapes, and you want each shape to be able to calculate its area. You can create an abstract base class Shape that defines a method area but doesn't implement it. Each specific shape (like Circle or Rectangle) will implement the area method.

Here's a simple example in Python:

```python
## copilot example
from abc import ABC, abstractmethod

# Abstract Base Class (ABC) module provides tools for defining abstract base classes
class Shape(ABC):
    # Abstract method that must be implemented by any subclass
    @abstractmethod
    def area(self):
        pass

# Subclass of Shape representing a Circle
class Circle(Shape):
    def __init__(self, radius):
        # Initialize the Circle with a radius
        self.radius = radius

    def area(self):
        # Calculate the area of the Circle using the formula Ï€r^2
        return 3.14 * self.radius ** 2

# Subclass of Shape representing a Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        # Initialize the Rectangle with width and height
        self.width = width
        self.height = height

    def area(self):
        # Calculate the area of the Rectangle using the formula width * height
        return self.width * self.height

# Creating an object of the Circle class with radius 5
circle = Circle(5)
# Creating an object of the Rectangle class with width 4 and height 6
rectangle = Rectangle(4, 6)

# Calling the area method on the Circle object and printing the result
print(circle.area())  # Output: 78.5
# Calling the area method on the Rectangle object and printing the result
print(rectangle.area())  # Output: 24
```

Access modifiers define the visibility of variables and methods in a class. They control how the data can be accessed from different parts of the program.

##### Public:

Visibility: Accessible from anywhere in the program.

Example: A public variable can be accessed and modified by any other class.

##### Protected:

Visibility: Accessible within the same class, subclasses (derived classes), and classes in the same package.

Example: A protected variable can be accessed by a subclass, but not by unrelated classes.

##### Private:

Visibility: Accessible only within the same class.

Example: A private variable cannot be accessed directly from outside the class. You need getter and setter methods to access and modify it.

Imagine you have a box (class) with a lock (private variable). Only you (the class itself) can open the lock directly. If someone else (another class) wants to see what's inside or put something in, they need to ask you (use getter and setter methods).

- Public: The box is open for everyone to see and use.
- Protected: Only your family (subclasses) and close friends (same package) can see and use the box.
- Private: Only you can see and use the box directly.

##### What is a cursor? 
Itâ€™s like a remote control for your database. It helps you run commands (like adding or retrieving data) and get results back.

##### Why use it? 
Without a cursor, you canâ€™t tell the database what to do from your Python code. Itâ€™s the link between your program and the database.

- You create a connection to the database (like opening the database).
- Then, you create a cursor (your helper).

```python
def reverse_vowels(word):
    '''
    returns the worde with reversed vowels in a given word
    '''
    word_lst=list(word)
    start=0
    end=len(word_lst)-1
    vowels='aeiouAEIOU'
    while start<end:
        if word_lst[start] in vowels and word_lst[end] in vowels:
            word_lst[start],word_lst[end]=word_lst[end],word_lst[start]
            start+=1
            end-=1
        elif word_lst[start] not in vowels:
            start+=1
        elif word_lst[end] not in vowels:
            end-=1
        
    return ''.join(word_lst)
```

```python
result=reverse_vowels("")
print(result)
result=reverse_vowels("leetcode")
print(result)
```

```python
def words_exactly_once(s1,s2):
    '''
    Return words that appear exactly once in both sentences
    '''
    lst_s1=s1.split(' ')
    lst_s2=s2.split(' ')
    final_lst=[]
    for i in lst_s1:
        if lst_s1.count(i)<2:
            if i not in lst_s2 and i not in final_lst:
                final_lst.append(i)
    for j in lst_s2:
        if lst_s1.count(i)<2:
            if j not in lst_s1 and j not in final_lst:
                final_lst.append(j)

    return final_lst
```
